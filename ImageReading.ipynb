```python
img = cv2.imread('puppy.jpg')
print(type(img))
print(img.shape)
```

```python
plt.imshow(img)
```

```python
img_color = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_color)
```

```python
img_grey = cv2.imread('puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_grey, cmap='gray')
```

```python
img = img_color
```

```python
plt.imshow(img)
```

```python
w_ratio = 0.25
h_ratio = 0.25
img_resize = cv2.resize(img, (0,0), img, w_ratio, h_ratio)
plt.imshow(img_resize)
```

```python
w_ratio = 0.30
h_ratio = 0.25
img_resize = cv2.resize(img, (0,0), img, w_ratio, h_ratio)
plt.imshow(img_resize)
```

```python
img_flip = cv2.flip(img, 1)
plt.imshow(img_flip)
```

```python
img_flip = cv2.flip(img, -1)
plt.imshow(img_flip)
```

```python
cv2.imwrite("totally_new_puppy.png", img_flip)
```
